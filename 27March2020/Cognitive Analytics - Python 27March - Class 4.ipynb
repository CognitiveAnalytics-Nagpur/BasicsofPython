```python
student_details = {'stu_id':[1,2,3,4,5],
                  'stu_name':['A','B','C','D','E'],
                  'Region':['N','S','E','W','W']}
sales_details = {'stu_id':[1,2,3,4,6],
                'Sub':['Py','R','Tab','SQL','Py'],
                'sales':[250, 125, 200, 200, 250]}
```

```python
sales_details = pd.DataFrame(sales_details)
student_details = pd.DataFrame(student_details)
```

```python
sales_details
```

```python
student_details
```

```python
pd.merge(student_details, sales_details, left_on='stu_id', right_on='stu_id',how='outer')
```

```python
pd.merge(student_details, sales_details, on='stu_id', how='left')
```
